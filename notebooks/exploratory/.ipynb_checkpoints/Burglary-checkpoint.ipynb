# Burglary
## Imports

```python
import pandas as pd
import bokeh as bk
import matplotlib.pyplot as plt
plt.style.use('ggplot')
# from mpl_toolkits.basemap import Basemap
```

## Opening data file

```python
crimes = pd.read_csv("../../data/crimes.csv")
```

## Keeping only burglary information

```python
crimes = crimes[crimes['CATEGORY'] == 'BURGLARY'].drop('Unnamed: 0', axis=1)
```

```python
crimes.head()
```

# Evolution by date

```python
# converting to date format
crimes['occ_date'] = pd.to_datetime(crimes['occ_date'], format='%Y-%m-%d')
```

```python
# dropping coordinates
evolution = crimes[['CATEGORY','occ_date']]
evolution.head()
```

## Sorting by date

```python
# sorting by date
evolution = evolution.sort_values(by='occ_date')
evolution.head()
```

### Adding date to index

```python
evolution.index = evolution['occ_date']
```

### Grouping by day

```python
# grouping and counting crimes on each day
counts_day =evolution.groupby(['occ_date']).agg(['count'])
counts_day.head()
```

## Plotting by day

```python
counts_day.plot(figsize=(150,10))
plt.savefig('../../plots/burglary_day_plot.png')
plt.show()
```

### Grouping by month

```python
counts_month = evolution.groupby(pd.TimeGrouper(freq='M')).agg(['count'])
counts_month = counts_month.drop('occ_date', axis=1)
counts_month.head()
#evolution = evolution.groupby([evolution.index.month])
#evolution.head()
```

```python
counts_month.plot(figsize=(100,10))
plt.savefig('../../plots/burglary_month_plot.png')
plt.show()
```
